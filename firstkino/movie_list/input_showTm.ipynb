### movie_info_detail에서 상영시간(showTm)을 읽어와서 DB에 넣는 작업
##### 작업자: 이종택

```python
#패키지 준비

import pymysql
import pandas as pd
```

```python
# 파일 불러오기 및 null값 0으로 채우기

df = pd.read_csv('../data-files/movie_info_result.csv', encoding='utf-8')
df = df.fillna(0)
```

```python
# 데이터 전처리 중 발생한 불필요한 값 제거(showTm컬럼 안에 'showTm'이 값으로 들어가 있는 문제)
# 데이터 정제하여(str -> float -> int, DataFrame -> list) DB에 넣기 전 하나의 리스트에 모두 넣는 작업

s = []

df = df.drop(df.loc[df['showTm'] == 'showTm'].index)
df['showTm'] = df['showTm'].astype('float').astype('int64')
processed_df = df[['movieCd', 'showTm']]
processed_df = processed_df.values.tolist()
s.append(processed_df)
```

```python
# DB접속 및 movieCd가 일치하는 행에 상영시간(showTm) 넣는 코드

conn = pymysql.connect(host="localhost",
                       database="firstkino",
                       user="kdigital",
                       password="mysql",
                       charset="utf8")

not_found_list = []
with conn.cursor() as cursor:
    for idx, post in enumerate(s[0]): # enumerate() 함수는 순서형 자료형의 인덱스와 값을 전달하는 기능하는 함수
        print(post[0], post[1])
        cursor.execute("SELECT showTm FROM movie WHERE movieCd=%s", (post[0]))
        selected_showtime = cursor.fetchall() # 위 코드를 실행하면 SELECT문을 실행하여 가져온 데이터가 selected_scores에 저장
        if len(selected_showtime) == 0:
            not_found_list.append(post)
        else:
            showtime =  selected_showtime[0][0] if selected_showtime[0][0] != None else post[1]
            print(type(showtime))
            cursor.execute("UPDATE movie set showTm = %s WHERE movieCd = %s", (showtime, post[0]))

conn.commit()

conn.close()
```
