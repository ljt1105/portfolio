```python
# 1. import 설정

import csv
import requests
import pandas as pd
from datetime import timedelta, datetime
import json
```

```python
# 2. movie_list_result.csv파일에서 영화코드 읽어오기

with open('./data-files/movie_list_result.csv', newline='', encoding="utf-8") as f:
    reader = csv.DictReader(f)

    # 영화코드를 담을 빈 리스트 생성
    movieCd_lists = []

    # 영화코드 한줄씩 읽어 리스트에 담기
    for row in reader:
        movieCd_lists.append(row['movieCd'])

        # 요청 수 브레이크
        # if len(movieCd_lists) == 10:
        #     break

print(len(movieCd_lists))
```

```python
# 2. 정보를 넣을 딕셔너리

# result = {}
```

```python
movie_info_detail=[]
# 3. 딕셔너리에 담은 movieCd로 영화 상세정보 요청
for i in range(6000,8999):

    # 3-1. url에 들어갈 키 요청
    key = '88cdf3b09edc68dac124e72da3c2a9a7'
    movieCd = movieCd_lists

    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={0}&movieCd={1}'.format(key, movieCd[i])

    # 2-2. 요청 데이터 저장
    
    response = requests.get(url).json()

    # movie_info_detail.append(json.loads(response)['movieInfoResult'])
    movie_info_detail.append(response['movieInfoResult']["movieInfo"])


movie_info_detail_df = pd.DataFrame(movie_info_detail)
```

```python
type(movie_info_detail_df["nations"][0])
movie_info_detail_df.head(1)
```

```python
movie_info_detail_df.head()

error_list = []
def extract_values(values, key):
    result = []
    try:
        for value in values:
            result.append(value[key])
    except:
        error_list.append(values)
    return result

movie_info_detail_df['nationsNm'] = movie_info_detail_df["nations"].map(lambda x : extract_values(x, 'nationNm'))
movie_info_detail_df['genreNm'] = movie_info_detail_df["genres"].map(lambda x : extract_values(x, 'genreNm'))
movie_info_detail_df['directors_ko'] = movie_info_detail_df["directors"].map(lambda x : extract_values(x, 'peopleNm'))
movie_info_detail_df['directors_en'] = movie_info_detail_df["directors"].map(lambda x : extract_values(x, 'peopleNmEn'))
movie_info_detail_df['actors_ko'] = movie_info_detail_df["actors"].map(lambda x : extract_values(x, 'peopleNm'))
movie_info_detail_df['actors_en'] = movie_info_detail_df["actors"].map(lambda x : extract_values(x, 'peopleNmEn'))
movie_info_detail_df['cast'] = movie_info_detail_df["actors"].map(lambda x : extract_values(x, 'cast'))
movie_info_detail_df['castEn'] = movie_info_detail_df["actors"].map(lambda x : extract_values(x, 'castEn'))
movie_info_detail_df['watchGradeNm'] = movie_info_detail_df["audits"].map(lambda x : extract_values(x, 'watchGradeNm'))
```
