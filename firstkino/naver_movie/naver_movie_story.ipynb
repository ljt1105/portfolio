```python
#패키지 준비

import requests
from bs4 import BeautifulSoup
from bs4.element import NavigableString
import time
import pandas as pd
```

```python
#영화 줄거리를 크롤링하기 전 각 영화페이지의 주소를 네이버 영화 평점 순위 순서대로 가져오는 작업

url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20211122&page={0}"
page = 1
movieNm = []
movieAddress = []

while True:
    response = requests.get(url.format(page))
    # print( response.status_code )
    soup = BeautifulSoup(response.text)
    movies = soup.select("#old_content .list_ranking tbody tr")

    if movies is None or len(movies) == 0:
        break
    
    # print(len(movies))

    for movie in movies:
        if len(movie.select('td')) == 1:
            continue

        title = movie.select_one('.title .tit5 a').get_text()
        address = movie.select_one('.title .tit5 a').get('href')

        #가져온 영화명과 웹페이지 주소를 행렬에 저장
        movieNm.append(title)
        movieAddress.append(address)
        
    page += 1
    time.sleep(1)

    if page == 41:
        break

#테스트
for movie in movies:
    print(movie)
```

```python
#영화명과 영화페이지 주소를 데이터프레임 형식으로 저장
df = pd.DataFrame({'movieNm':movieNm, 'movieAddress':movieAddress})
print(df.head())
print(df.info())
```
